```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import polars as pl
```

```python
# Importing the dataset

url = 'https://drive.google.com/uc?export=download&id=1z8mSbWzBY2bH2ktuGA6OK-3kYB9gOawv'
df = pd.read_csv(url)
```

```python
# finding duplicates

duplicates = df[df.duplicated()]
print("Duplicates:")
print(duplicates)
```

```python
# finding missing values

missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# finding null values

null_values = df.isnull().sum()
print("Null Values:")
print(null_values)
```

```python
#  Filling null value with mean of bmi column

df.fillna(df['bmi'].mean())
```

```python
# Finding columns with constant values (irrelevant data)

constant_columns = [col for col in df.columns if df[col].nunique() == 1]
print("Columns with constant values (possibly irrelevant):")
print(constant_columns)
```

```python
# Finding data type mismatches

expected_dtypes = {
    'id': int,
    'gender': object,
    'age': float,
    'hypertension': int,
    'heart_disease': int,
    'ever_married': object,
    'work_type': object,
    'Residence_type': object,
    'avg_glucose_level': float,
    'bmi': float,
    'smoking_status': object,
    'stroke': int
}
dtype_mismatches = {col: (df[col].dtype, expected) for col, expected in expected_dtypes.items() if df[col].dtype != expected}
print("Data Type Mismatches:")
print(dtype_mismatches)
```

```python
# Data Formatting (Standardizing string columns to lowercase)

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower()

print("Data Formatting Applied:")
print(df.head(10))
```

```python
# Using the Box plot for checking the outliers

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, color="orange")
plt.show()
```

```python
# Using IQR method for detecting outliers

def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers = {col: find_outliers_iqr(df, col) for col in df.select_dtypes(include=np.number).columns}

print("Outliers:")
for col, outlier_df in outliers.items():
    print(f"{col} Outliers:")
    print(outlier_df)
```

```python
# Finding Bias and mitigating it

# Check for class imbalance in the target variable 'stroke'
class_counts = df['stroke'].value_counts()
print("Class Distribution for 'stroke':")
print(class_counts)

# Mitigate class imbalance by resampling (undersampling or oversampling)
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df[df.stroke == 0]
df_minority = df[df.stroke == 1]

# Upsample minority class
df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=123)

# Combine majority class with upsampled minority class
df_resampled = pd.concat([df_majority, df_minority_upsampled])

print("Class Distribution after Resampling:")
print(df_resampled['stroke'].value_counts())
```
