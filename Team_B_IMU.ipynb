```python
def plot_distributions(data, columns, rows=2, cols=2):
    fig, axes = plt.subplots(rows, cols, figsize=(14, 8))
    axes = axes.flatten()

    for i, column in enumerate(columns):
        if i < rows * cols:
            sns.histplot(data[column], kde=True, bins=30, ax=axes[i])
            axes[i].set_title(f'Distribution of {column}')
            axes[i].set_xlabel(column)
            axes[i].set_ylabel('Frequency')
        else:
            axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
# Function to extract statistical features
def extract_features(df):
    features = pd.DataFrame()

    features['mean'] = df.mean()
    features['std'] = df.std()
    features['min'] = df.min()
    features['max'] = df.max()
    features['median'] = df.median()
    features['skew'] = df.skew()
    features['kurtosis'] = df.kurtosis()

    return features
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

def random_forest_feature_selection(data, n_features=None, threshold=None, random_state=42, n_estimators=50):

    # Scale the data
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data)

    # Initialize and train the model
    model = RandomForestClassifier(n_estimators=n_estimators, random_state=random_state, n_jobs=-1)
    model.fit(scaled_data, np.random.randint(0, 2, size=scaled_data.shape[0]))

    # Get feature importances
    importances = model.feature_importances_

    feature_importances = pd.DataFrame({'Feature': data.columns, 'Importance': importances})
    feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

    # Plot feature importances
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feature_importances)
    plt.title('Feature Importances from Random Forest')
    plt.show()

    # Select features based on threshold or top n features
    if threshold:
        selected_features = feature_importances[feature_importances['Importance'] >= threshold]['Feature'].tolist()
    elif n_features:
        selected_features = feature_importances['Feature'].tolist()[:n_features]
    else:
        selected_features = feature_importances['Feature'].tolist()

    return data[selected_features], selected_features

n_features_to_select = 5
importance_threshold = 0.05

# Assume Vicon_comb_df and Kinect_comb_df are already defined
Vicon_selected_features_df, Vicon_selected_features = random_forest_feature_selection(Vicon_comb_df, n_features=n_features_to_select)
print("Selected features for Vicon data after Random Forest feature selection:", Vicon_selected_features)

Kinect_selected_features_df, Kinect_selected_features = random_forest_feature_selection(Kinect_comb_df, n_features=n_features_to_select)
print("Selected features for Kinect data after Random Forest feature selection:", Kinect_selected_features)
```

**3. PCA**

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Define a function to compute reconstruction error using PCA
def pca_reconstruction_error(data, n_components):
    pca = PCA(n_components=n_components)
    transformed_data = pca.fit_transform(data)
    reconstructed_data = pca.inverse_transform(transformed_data)
    error = np.mean((data - reconstructed_data) ** 2)
    return error

# Standardize the data
scaler = StandardScaler()
Vicon_scaled = scaler.fit_transform(Vicon_comb_df)
Kinect_scaled = scaler.fit_transform(Kinect_comb_df)

# Backward elimination for Vicon data
Vicon_selected_features = list(range(Vicon_scaled.shape[1]))
min_error = pca_reconstruction_error(Vicon_scaled[:, Vicon_selected_features], n_components=len(Vicon_selected_features))

while len(Vicon_selected_features) > 1:
    errors = []
    for feature in Vicon_selected_features:
        current_features = [f for f in Vicon_selected_features if f != feature]
        error = pca_reconstruction_error(Vicon_scaled[:, current_features], n_components=len(current_features))
        errors.append((feature, error))

    feature, error = min(errors, key=lambda x: x[1])
    if error <= min_error:
        min_error = error
        Vicon_selected_features.remove(feature)
    else:
        break

# Backward elimination for Kinect data
Kinect_selected_features = list(range(Kinect_scaled.shape[1]))
min_error = pca_reconstruction_error(Kinect_scaled[:, Kinect_selected_features], n_components=len(Kinect_selected_features))

while len(Kinect_selected_features) > 1:
    errors = []
    for feature in Kinect_selected_features:
        current_features = [f for f in Kinect_selected_features if f != feature]
        error = pca_reconstruction_error(Kinect_scaled[:, current_features], n_components=len(current_features))
        errors.append((feature, error))

    feature, error = min(errors, key=lambda x: x[1])
    if error <= min_error:
        min_error = error
        Kinect_selected_features.remove(feature)
    else:
        break

# Print selected features
print("Selected features for Vicon data:", Vicon_selected_features)
print("Selected features for Kinect data:", Kinect_selected_features)
```

**7.BACKWARD FEATURE ELIMINATION**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Assuming Vicon_comb_df and Kinect_comb_df are your dataframes
# Standardize the data
scaler = StandardScaler()
Vicon_comb_df_scaled = scaler.fit_transform(Vicon_comb_df)
Kinect_comb_df_scaled = scaler.fit_transform(Kinect_comb_df)

# Convert scaled data back to DataFrame
Vicon_comb_df_scaled = pd.DataFrame(Vicon_comb_df_scaled, columns=Vicon_comb_df.columns)
Kinect_comb_df_scaled = pd.DataFrame(Kinect_comb_df_scaled, columns=Kinect_comb_df.columns)

# Backward feature elimination using SequentialFeatureSelector
def backward_elimination(df):
    # Using Linear Regression model to perform backward elimination
    model = LinearRegression()
    sfs = SequentialFeatureSelector(model, n_features_to_select='auto', direction='backward')
    sfs.fit(df, df)
    selected_features = df.columns[sfs.get_support()]
    return selected_features

# Apply backward elimination to Vicon_comb_df
selected_features_vicon_backward = backward_elimination(Vicon_comb_df_scaled)
print("Selected features in Vicon data after backward elimination:")
print(selected_features_vicon_backward)

# Apply backward elimination to Kinect_comb_df
selected_features_kinect_backward = backward_elimination(Kinect_comb_df_scaled)
print("Selected features in Kinect data after backward elimination:")
print(selected_features_kinect_backward)

# Function to visualize clusters using PCA and KMeans
def plot_kmeans_clusters(df, selected_features, title, sample_size=500):
    # Randomly sample data points to avoid congestion
    sampled_df = df[selected_features].sample(n=sample_size, random_state=42)

    # Reduce dimensionality for visualization using PCA
    pca = PCA(n_components=2, random_state=42)
    pca_results = pca.fit_transform(sampled_df)

    # Apply KMeans clustering
    kmeans = KMeans(n_clusters=3, random_state=42)
    clusters = kmeans.fit_predict(sampled_df)

    plt.figure(figsize=(12, 6))
    plt.scatter(pca_results[:, 0], pca_results[:, 1], c=clusters, cmap='viridis', label='Data points')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
    plt.title(f'KMeans Clusters: {title}')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend()
    plt.show()

# Plot KMeans clusters for selected features in Vicon data
plot_kmeans_clusters(Vicon_comb_df_scaled, selected_features_vicon_backward, 'Vicon (Backward Elimination)')

# Plot KMeans clusters for selected features in Kinect data
plot_kmeans_clusters(Kinect_comb_df_scaled, selected_features_kinect_backward, 'Kinect (Backward Elimination)')


# Plot KMeans clusters for selected features in Vicon data
plot_kmeans_clusters(Vicon_comb_df_scaled, selected_features_vicon_backward, 'Vicon (Backward Elimination)')

# Plot KMeans clusters for selected features in Kinect data
plot_kmeans_clusters(Kinect_comb_df_scaled, selected_features_kinect_backward, 'Kinect (Backward Elimination)')
```

**INSIGHTS:**
* The clustering analysis reveals three distinct movement patterns captured by the sensors. The dense purple cluster likely represents a common movement pattern seen across many exercises, indicating repetitive or standardized motions.

* The more spread-out teal cluster suggests variability, possibly due to different types of exercises or variations within the same exercise.

* The sparse yellow cluster might indicate rare or outlier movements, which could be errors, unique exercises, or transitions between exercises.
